```python
import io
import pandas as pd
```

```python
# читаем данные из файла
import csv
  
with open('light.csv', 'r') as read_obj:
    dataframe= pd.read_csv(read_obj)
    print(dataframe)
```

```python
# задаем пороговую освещенность и начальное состояние света
threshold = 35
light_on = False
```

```python

# проходимся по каждой строке данных
for ind, row in dataframe.iterrows():
    # получаем текущее значение освещенности
    light = float(row[0])
    # если освещенность ниже порога
    if light < threshold:
        if light_on:
          print(f'Яркость: {light}. Выключаем свет')
          light_on = False
    # если освещенность выше порога
    elif light >= threshold :
        if not light_on:
          print(f'Яркость: {light}. Включаем свет')
          light_on = True
```
